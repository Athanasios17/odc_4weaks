```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\yostina\Downloads\archive (1).zip")
df.head(10)
```

```python
df.duplicated().sum()
```

```python
df['Nationality'] = df['Nationality'].fillna('unspecified')
```

```python
df['Age'] = df['Age'].fillna(df.mean)
df['Age']
df['Headed goals']=df['Headed goals'].fillna(0)
```

```python
df=df.fillna(0)
```

```python
df.isna().mean()
```

```python
sns.boxplot(x=df['Red cards'])
plt.show()
```

```python
HeadGoals=df.groupby('Name')['Headed goals'].sum().sort_values(ascending=False)
HeadGoals.head(5).plot(kind='bar', title='GOALS BY HEAD',xlabel='Name')
```

```python
Loseclub=df.groupby('Club')['Losses'].sum().sort_values(ascending=False)
Loseclub.head(6).plot(kind='barh',title='NUMBER OF LOSE ')
```

```python
Nat=df['Nationality'].value_counts().sort_values(ascending=False)
Nat.head(10).plot(kind='pie',title='Nationality')
```

```python
Shooting=df.groupby('Name')['Shooting accuracy %'].value_counts().sort_values(ascending=False)
Shooting.head(15).plot(kind='pie',title='Shooting Accuracy')
```

```python
Miss=df.groupby('Name')['Big chances missed'].sum().sort_values(ascending=False)
Miss.head(20).plot(kind='area')
```

```python
plt.figure(figsize=(20, 12))
sns.scatterplot(x='Age',y='Goals per match',data=df)
plt.show()
```
